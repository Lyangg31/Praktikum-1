## Pie Chart

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')

stock = ['Cheese', 'Bread', 'Ice Cream', 'Noodles', 'Egg']
sold = [10, 50, 45, 15, 37]

ax.pie(sold, labels=stock, autopct='%1.2f%%')

plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data untuk tiga kategori
akomodasi = [2000, 1100, 4180, 5800, 6420, 980]
kamar = [1503, 10732, 3555, 5206, 2187, 4731, 522]
kasur = [9257, 16415, 6045, 7832, 2885, 8080, 786]

# Membuat figure dan axis
fig, ax = plt.subplots(figsize=(8,6))  # Ukuran lebih jelas

# Membuat boxplot dengan berbagai pengaturan
box = ax.boxplot([akomodasi, kamar, kasur],
                 vert=False,  # Membuat boxplot horizontal
                 showmeans=True,  # Menampilkan nilai mean
                 meanline=True,  # Menampilkan mean sebagai garis
                 tick_labels=['Akomodasi', 'Kamar', 'Kasur'],  # Perbaikan: 'labels' -> 'tick_labels'
                 patch_artist=True,  # Memberikan warna pada boxplot
                 medianprops={'linewidth': 2, 'color': 'blue'},  # Warna garis median
                 meanprops={'linewidth': 2, 'color': 'red'})  # Warna garis mean

# Menambahkan warna untuk setiap box
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Menyesuaikan aspek agar tidak terjadi peringatan
ax.set_aspect('auto')

# Menampilkan plot
plt.show()
```

```python
import matplotlib.pyplot as plt

#Data nilai tugas, UTS, dan UAS 
tugas = [58,80,65,79,82]
uts = [78,80,85,59,92]
uas = [62,79,87,77,84]

#membuat figure dan axis
fig, ax = plt.subplots()

#scatter plot hubungan tugas dengan UTS
ax.scatter(tugas,uts, color='red', label='Tugas vs UTS')

#scatter plot hubungan tugas dengan UAS
ax.scatter(tugas,uas, color='blue', label='Tugas vs UAS')

#Menambahkan label sumbu dengan judul
tempX = 'Tugas'
tempY = 'Nilai'
ax.set_xlabel(tempX)
ax.set_ylabel(tempY)
ax.set_title('Scatter Plot')

#Menampilkan legenda
ax.legend()

#Menampilkan plot
plt.show()
```
